```python
#Crear la base de datos
def createDB():
  conn = sql.connect("Ventas.db")
  conn.commit()
  conn.close()
  print("Cree la base de datos")
```

```python
#Crear la entidad (tabla)
def tableDB():
  conn = sql.connect("Ventas.db")
  cursor = conn.cursor()
  cursor.execute("""CREATE TABLE ventas(
    Fecha_Venta datetime,
    Año INT,
    Semana_Venta INT,
    Desc_Producto INT,
    Presentacion TEXT,
    Tipo_Producto TEXT,
    Cant_Vendida TEXT,
    Precio_Venta INT,
    Venta_Total INT
  );
  """)
  conn.commit()
  conn.close()
  print("Cree la tabla")
```

```python
df_ventas = pd.read_csv("/content/drive/MyDrive/Gestion y almacenamiento datos/data_ventas.csv", header = 0)
```

```python
#Convierto mi dataframe de Pandas a base de datos de SQL
conn = sql.connect('Ventas.db')
df_ventas.to_sql('ventas', conn, if_exists="replace", index=False)
conn.close()
```

```python
#Consulto los datos cargados
conn = sql.connect('Ventas.db')
print("Me conecté a la base de datos")
cursor = conn.cursor()
query = f"SELECT * FROM ventas"
cursor.execute(query)
datos = cursor.fetchall()
conn.commit()
print(datos)
conn.close()
```

```python
#Eliminar la columna "concat"
conn = sql.connect('Ventas.db')
cursor = conn.cursor()
query = f"ALTER TABLE ventas DROP COLUMN Concat;"
cursor.execute(query)
conn.commit()
conn.close()
print("Borró la columna")
```

```python
#Verifico que la columna haya quedado eliminada
conn = sql.connect('Ventas.db')
cursor = conn.cursor()
query = f"SELECT * FROM ventas"
cursor.execute(query)
datos = cursor.fetchall()
conn.commit()
print(datos)
conn.close()
```
