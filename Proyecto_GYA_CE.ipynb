```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Cargar el archivo Titanic-Dartaset
df_ventas = pd.read_excel("/content/drive/MyDrive/Gestion y almacenamiento datos/Ventas_proyectoGAD.xlsx", header = 0)
df_compras = pd.read_excel("/content/drive/MyDrive/Gestion y almacenamiento datos/Compras_proyectoGAD.xlsx", header = 0)
```

```python
df_ventas = df_ventas.drop(['Concat'],axis=1)
```

```python
df_compras = df_compras.drop(['Concat','Margen $','Total Venta','Total Vta Prom','Venta Neta','Margen Total $'],axis=1)
```

```python
df_ventas['Fecha Venta'] = df_ventas['Fecha Venta'].astype("datetime64")
df_compras['Fecha Entrega'] = df_compras['Fecha Entrega'].astype("datetime64")
```

```python
#Ventas por semana-año
ventas_semana = df_ventas.groupby(['Año','Semana Venta'])['Venta Total'].sum()
fecha_venta = [f"{a}-{s:02d}" for a, s in ventas_semana.index]
```

```python
#Grafica de ventas por semana
plt.figure(figsize=(15,6))
plt.plot(fecha_venta,ventas_semana)
plt.xlabel("Fecha")
plt.ylabel("$")
plt.xticks( rotation = 90)
plt.title("Ventas por semana")
```

Se observan ventas irregulares samana a semana, esto obedece a que el negocio se encuentra en etapa de estabilización, sin embargo, se observa de manera positiva que la mayoria de las semanas presentan ventas superiores a los $400.000.
Este panorama refuerza la idea de hacer un seguimiento a detalle del comportamiento de las ventas

```python
venta_totales = df_ventas['Venta Total'].sum()
compras_totales = df_compras['Costo Total'].sum()

plt.subplots(figsize=(6,4))
metricas = ['Ventas Totales', 'Compras Totales']
valores = [venta_totales, compras_totales]
bar_colors = ['blue', 'red']

plt.bar(metricas, valores,color=bar_colors)

plt.title('Ventas vs Compras')
plt.show()
```

Esta gráfica muestra un panorama positivo ya que indica que en lo corrido del ejercici se esta generando una rentabilidad positiva (las ventas son superiores a las compras)

```python
#Compras por semana-año
compras_semana = df_compras.groupby(['Año','Semana'])['Costo Total'].sum()
fecha_compra = [f"{a}-{s:02d}" for a, s in ventas_semana.index]
```

```python
plt.figure(figsize=(15,6))
plt.plot(fecha_venta,ventas_semana,'b-',label="Ventas")
plt.plot(fecha_compra,compras_semana,'r--',label="Compras")
plt.legend(loc="upper right")
plt.xlabel("Fecha")
plt.ylabel("$")
plt.xticks( rotation = 90)
plt.title("Ventas y compras por semana")
```

Este tipo de gráficas donde se comparan las compras y ventas de manera tendencial nos deja ver los momentos de abastecimiento, y ayudan a planear de una manera adecuada las futuras compras con el fin de general la mayor rentabilidad

```python
#Rentabilidad total
rent_total = round((venta_totales/compras_totales-1)*100,1)
rent_total
```

```python
#Rentabilidad por semana
rentabilidad = round((ventas_semana/compras_semana-1)*100,1)
rentabilidad
```

```python
#Grafico de la Rentabilidad por semana

f_rent = [f"{a}-{s:02d}" for a, s in rentabilidad.index]
plt.figure(figsize=(15,6))
plt.plot(f_rent,rentabilidad)
plt.xlabel("Periodo")
plt.ylabel("%")
plt.xticks( rotation = 90)
plt.title("Rentabilidad por semana. Rentabilidad Total: "+str(rent_total)+"%")
```

Gráfica de Rentabilidad: Es importante destacar que, sin bien, al principio del ejercicio es aceptable tener márgenes negativos, al pasar los periodos lo que se espera es que el margen sea positivo, o cero, lo que indica que se esta planeando de manera acertada el proceso de abastecimiento de inventario.

```python
#Categorias más vendidas
venta_categoria = round((df_ventas.groupby(['Tipo Producto'])['Venta Total'].sum())/df_ventas['Venta Total'].sum()*100)
venta_mostrar = venta_categoria.sort_values(ascending=False)
```

```python
plt.pie(venta_mostrar, labels=venta_mostrar.index, rotatelabels=True, autopct='%1.0f%%',pctdistance = 0.9,
        wedgeprops = {"linewidth": 1, "edgecolor": "white"})
plt.title("Porcentaje de ventas por categoria")
plt.show
```

Esta gráfica es importante y sirve como punto de partida para saber qué tipo de productos no pueden faltar en el portafolio de la tienda y cuales podrian ser retirados. Esto con el fin, nuevamente, de generar una mayor rentabilidad al negocio

```python
#Los productos más vendidos
productos_mas_vendidos = df_ventas.groupby(['Desc Producto','Presentacion'])['Venta Total'].sum()
productos_mas_vendidos.sort_values(ascending=False)
```
